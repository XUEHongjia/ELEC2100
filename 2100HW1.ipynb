```python
import numpy as np
import matplotlib.pyplot as plt

num = 100
t1 = np.arange(-5,-2, 1/num)
t2 = np.arange(-2, 0, 1/num)
t3 = np.arange( 0, 2, 1/num)
t4 = np.arange( 2, 5, 1/num)
t = t1
t = np.append(t,t2)
t = np.append(t,t3)
t = np.append(t,t4)

x1 = np.zeros(num*3)
x1 = np.append( x1, np.arange( 2, 0, -1/num ) )
x1 = np.append( x1, - np.ones( 2*num ) )
x1 = np.append( x1, np.zeros(num*3) )

x1_ = np.zeros(10*num)
for i in range ( 10*num ):
    x1_[i] = x1[-i-1]

even = ( x1 + x1_ )/2
odd = ( x1 - x1_ )/2

plt.title("x1(t)_even_part -- t")
plt.xlabel("time(sec) - label")
plt.ylabel("even_part_of_x1(t) - label")
plt.plot( t, even )

plt.grid( )

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


plt.title("x1(t)_odd_part -- t")
plt.xlabel("time(sec) - label")
plt.ylabel("odd_part_of_x1(t) - label")

plt.plot(t, odd)

plt.grid()

plt.show()
```
