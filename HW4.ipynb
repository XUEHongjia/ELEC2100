```python
# Q2
import numpy as np
import matplotlib.pyplot as plt
import math
PI = math.pi
NUM = 1000
t = np.arange(-0.25,0.25,1/NUM)
x = np.zeros(len(t))
for i in range(0,len(t)):
    x[i] = math.sin( 25*PI*t[i] )/(PI*t[i])
    
plt.title("")
plt.xlabel("t")
plt.ylabel("x(t)")
plt.plot( t, x )

plt.grid(0.04)
print ( "" )
plt.show()
```

```python
t = np.arange(-30*PI,30*PI,1/NUM)
x = np.zeros( len(t) )
for i in range( 0, len(t) ):
    if math.fabs( 30*PI*NUM - i ) < 25*PI*NUM:
        x[i] = 1
    
plt.title("")
plt.xlabel("omega")
plt.ylabel("magnitude spectrum of x(t)")
plt.plot( t, x )

plt.grid( )
print ( "" )
plt.show()
```

```python
#Q4 b
omega = np.arange(-PI,PI,0.01)
X = np.zeros( len( omega ) )
for i in range( 0, len( omega ) ):
    X[i] = math.sin( omega[i]*(2+0.5) )/( math.sin( 0.5*omega[i] ) )
    
plt.title("")
plt.xlabel("omega")
plt.ylabel(" DTFT ")
plt.plot( omega, X )

plt.grid( )
print ( "" )
plt.show()
```
