```python
# Q1
import numpy as np
import matplotlib.pyplot as plt
import math
t = np.arange(-7,8)
x = np.zeros(15)
PI = math.pi
x[7] = 1
x[7-2] = 3
x[7+2] = 3
x[7-5] = 1
x[7+5] = 1
    
plt.title("")
plt.xlabel("k")
plt.ylabel("magnitude of ak")
plt.scatter( t, x )

plt.grid( )
print ( "" )
plt.show()
```

```python
t1 = np.arange(-7,8)
x1 = np.zeros(15)
PI = math.pi
x1[7] = 0
x1[7-2] = PI/2
x1[7+2] = -PI/2
x1[7-5] = -PI/4
x1[7+5] = PI/4
    
plt.title("")
plt.xlabel("k")
plt.ylabel("phase of ak")
plt.scatter( t1, x1 )

plt.grid( )
print ( "" )
plt.show()
```

```python
t4 = np.arange(-15,16)
x4 = np.zeros(31)
for i in range(0,31):
    if (i-15) % 8 == 3 or (i-15) % 8 == 4 or (i-15) % 8 == 5:
        x4[i] = 0
    else:
        x4[i] = 1
        
plt.title("")
plt.xlabel("n")
plt.ylabel("z[n]")
plt.scatter( t4, x4 )

plt.grid( )
print ( "" )
plt.show()
```
