```python
# Q1
import numpy as np
import matplotlib.pyplot as plt
import math
t = np.arange(0,20)
x = np.zeros(20)
PI = math.pi
for i in range(20):
    x[i] = math.cos( t[i]*2*PI/3 ) + math.cos( t[i]*4*PI/3 )
    
plt.title("")
plt.xlabel("")
plt.ylabel("")
plt.scatter( t, x )

plt.grid( )
print ( "" )
plt.show()
```

```python
#Q4.c
t = np.arange(-5,16)
x = np.zeros(21)
for i in range(4):
    x[i+5-1] = 2
    

y1 = np.zeros(21)
for i in range (21):
    if i < -3+5:
        y1[i] = 0
    elif i > 10+5:
        y1[i] = 0
    elif i > 4+5 :
        y1[i] = x[i-2] + x[i-8]
    else :
        y1[i] = x[i-2]
        
plt.scatter( t,y1 )
```

```python
#Q4.d
z1 = np.zeros(21)
for i in range (21):
    for n in range(i):
        z1[i] += y1[n]
        
plt.scatter(t,z1)
```
