```python
#Q1 c
import numpy as np
import matplotlib.pyplot as plt
import math
PI = math.pi
f = 20000
omega = np.arange(-2*PI*f*2, 2*PI*f*2+1, 2*PI*20)
P = np.zeros( len(omega) )
for i in range ( 0, len(P) ):
    if ( i % 1000 == 0 ):
        P[i] = 2*PI*f
    
plt.title("")
plt.xlabel("omega")
plt.ylabel("P")
plt.scatter( omega, P )

plt.grid()
print ( "" )
plt.show()
```

```python
#Q1 e
import numpy as np
import matplotlib.pyplot as plt
import math
PI = math.pi
f = 20000
omega = np.arange(-PI*3, PI*3, PI/100)
XP = np.zeros( len(omega) )
for i in range( len(XP) ):
    if ( abs( i-100 )<=20 or abs( i-300 )<=20 or abs( i-500 )<=20  ):
        XP[i] = 1
    
plt.title("")
plt.xlabel("omega")
plt.ylabel("XD")
plt.plot( omega, XP )

plt.grid()
print ( "" )
plt.show()
```

```python
#Q1 e
import numpy as np
import matplotlib.pyplot as plt
import math
PI = math.pi
f = 20000
omega = np.arange(-PI*3, PI*3, PI/100)
XP = np.zeros( len(omega) )
for i in range( len(XP) ):
    if ( abs( i-300 )<=20  ):
        XP[i] = (f - 500*abs( i-300 ))/f

    
plt.title("")
plt.xlabel("omega")
plt.ylabel("XD")
plt.plot( omega, XP )

plt.grid()
print ( "" )
plt.show()
```

```python
NUM = 10000
t = np.arange(-2*PI*NUM*5,2*PI*NUM*5+PI*NUM, 2*PI*NUM )
x = np.zeros(len(t))
for i in range(0,len(t)):
    if i == 5:
        x[i] = PI/2
    else:
        x[i] = 2*math.sin( PI/4*( i-5 ) )/(i-5)
    
plt.title("")
plt.xlabel("t")
plt.ylabel("x(t)")
plt.scatter( t, x )

plt.grid(0.04)
print ( "" )
plt.show()
```
